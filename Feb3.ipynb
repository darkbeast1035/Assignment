```python
#Q4 
#A generator function in python returns an iterator object. Instead of using return to send back a single value, generator functions use yield to produce a series of results over time. The function pauses its execution after yield, maintaining its state between iterations.

def test1():
    n=5
    c=1
    while(c<=n):
        yield c
        c+=1
for n in test1():
    print(n)
```

```python
#Q5
def prime():
    n=1000
    c=1
    while(c<=n):
        counter = 0
        for i in range(1,c+1):
            if c%i==0:
                counter+=1
        else:
            if counter==2:
                yield c
        c+=1
num_iterator = prime()
for _ in range(20):
    print(next(num_iterator))
```
