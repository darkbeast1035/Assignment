```python
#Q1
def password_check(password):
    upper_count, lower_count, num_count, spl_count = 0, 0, 0, 0
    for char in password:
        if ord(char) in range(48,58):
            num_count+=1
        elif ord(char) in range(65,91):
            upper_count+=1
        elif ord(char) in range(97,123):
            lower_count+=1
        elif (ord(char) in range(33,48)) or (ord(char) in range(58,65)) or (ord(char) in range(91,97)) or (ord(char) in range(123,127)):
            spl_count+=1

    if len(password)==10 and upper_count>=2 and lower_count>=2 and spl_count>=3 and num_count>=1:
        return "Valid Password"
    else:
        return "Invalid Password"
password_check("P@s*W0@d!1") #valid
password_check("PassW0@d!1") #invalid
```
