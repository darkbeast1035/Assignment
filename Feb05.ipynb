```python
#Q3
#__init__() is a constructor used to initialise attributes of an object
#for ex:-
#Student registered in pwskills. The student has attributes like name, number and student id.
#To inititialise those attributes specific to one student __init__() constructor is used

class Animal:

    def __init__(self,name,legs): #initialising attributes of an object
        self.name = name
        self.legs = legs

    def details(self):
        return self.name, self.legs

cat = Animal("Cat",4)
cat.details()
```
