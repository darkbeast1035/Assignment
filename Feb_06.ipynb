```python
#Q1
from functools import reduce
list1 = [1,2,3,4, [44,55,66, True], False, (34,56,78,89,34), {1,2,3,3,2,1}, {1:34, "key2": [55, 67, 78, 89], 4: (45,22, 61, 34)}, [56, 'data science'], 'Machine Learning']
def flatten_list(list1):
    flat_list = []
    for item in list1:
        if isinstance(item,list) or isinstance(item,tuple) or isinstance(item,set):
            flat_list.extend(flatten_list(item))
        elif isinstance(item, dict):
            flatten_list(list(item.keys()))
            flatten_list(list(item.values()))
        else: 
            flat_list.append(item)
    return flat_list

flatten_list_data = flatten_list(list1)
flatten_list_data = [i for i in flatten_list_data if type(i)==int ]

reduce(lambda a,b: a*b, flatten_list_data)
```
